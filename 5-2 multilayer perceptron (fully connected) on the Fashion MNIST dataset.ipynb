```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier, KerasRegressor
from keras import regularizers
import keras

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialization
digits = load_digits()
# print(digits.data.shape)
input_size = digits.data.shape[1]

X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target,random_state=1)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= np.max(X_train) # scaling between 0 and 1
X_test /= np.max(X_test)

num_classes = 10

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
np.max(X_train)
X_train.shape
```

```python
# Wrapper method
def make_model(optimizer="adam", hidden_size=128, kern_reg = 0.00001):
  model = Sequential([
      Dense(hidden_size, input_shape=(input_size,), kernel_regularizer = regularizers.l2(kern_reg) ),
      Activation('relu'),
      Dense(hidden_size ),
      Activation('relu'),
      Dense(10),
      Activation('softmax'), ])
  
  model.compile(optimizer=optimizer,loss="categorical_crossentropy", metrics=['accuracy'])
  return model


clf = KerasClassifier(make_model)             
param_grid = {'epochs': [1,3,10], 'kern_reg': [ 0.00001,0.0001, 0.001, 0.01, 10], 'hidden_size': [16, 32, 64,128]}

grid = GridSearchCV(clf, param_grid=param_grid)
grid.fit(X_train, y_train, verbose=0)
```

```python
print('Best score: ',grid.best_score_)
print('Best param: ',grid.best_params_)

score = grid.score(X_test, y_test)
print("test score: ", score)
```
